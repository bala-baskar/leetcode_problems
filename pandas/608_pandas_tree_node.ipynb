```python
%%writefile /content/leetcode_problems/pandas/p608_tree_node.py
'''
Link: https://leetcode.com/problems/tree-node/description/


Solution:
'''
import pytest
import pandas as pd

# Write function to test with pytest
def tree_node(tree: pd.DataFrame) -> pd.DataFrame:
  tree['type'] = None
  tree.loc[pd.isnull(tree['p_id']),'type'] = 'Root'
  tree.loc[(pd.isnull(tree['type'])) & (~tree['id'].isin(list(tree['p_id'].unique()))),'type'] = 'Leaf'
  tree.loc[pd.isnull(tree['type']),'type'] = 'Inner'
  return tree[['id','type']]

# Dictionary containing test cases
test_cases = {
    'case_1': {
        'input': pd.DataFrame({
            'id': ['1', '2', '3', '4', '5'],
            'p_id': [None, '1', '1', '2', '2']
        }),
        'output': pd.DataFrame({
            'id': ['1', '2', '3', '4', '5'],
            'type': ['Root', 'Inner', 'Leaf', 'Leaf', 'Leaf']
        })
    },
    'case_2': {
        'input': pd.DataFrame({
            'id': ['1'],
            'p_id': [None]
        }),
        'output': pd.DataFrame({
            'id': ['1'],
            'type': ['Root']
        })
    },
    'case_3': {
        'input': pd.DataFrame({
            'id': ['1','2','3','4','5','6','7'],
            'p_id': ['2',None,'2','3','4','5','5']
        }),
        'output': pd.DataFrame({
            'id': ['1','2','3','4','5','6','7'],
            'type': ['Leaf','Root','Inner','Inner','Inner','Leaf','Leaf']
        })
    }
}

# Test function using pytest
@pytest.mark.parametrize("test_name, case", test_cases.items())
def test_process_dataframe(test_name, case):
    input_df = case['input']
    expected_df = case['output']
    result_df = tree_node(input_df)
    assert type(result_df) == pd.DataFrame
    # pd.testing.assert_frame_equal(result_df, expected_df)
    assert result_df.equals(expected_df)
```
